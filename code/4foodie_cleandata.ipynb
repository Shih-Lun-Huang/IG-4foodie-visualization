```python
res = pd.DataFrame()
for i in range(len(df)):
    for j in range(len(df['mrt'][i])):
        if df['mrt'][i][j] != "":
            data = df.loc[i]
            data['new_mrt'] = df['mrt'][i][j]
            data['X'] = map_dict[df['mrt'][i][j]][0]
            data['Y'] = map_dict[df['mrt'][i][j]][1]
            res = res.append(data)
res.reset_index(drop=True,inplace=True)
```
